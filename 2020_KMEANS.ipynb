```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import math
from sklearn.cluster import KMeans
import tensorflow as tf
#from sklearn.datasets import make_blobs
#import pandas
```

```python
def dataShape(centerList, choice, X, Y):
    center = []
    for i in range(len(centerList)): #센터 점 리스트구조로 변형
        center.append([centerList[:,0][i], centerList[:,1][i]])
        
    #center점의
    for i in range(len(center)-1):
        for j in range(i+1, len(center)):
            if center[i][0] > center[j][0]:
                tmp = center[i]
                center[i] = center[j]
                center[j] = tmp
            
    train = []
    for i in range(len(center)):
        train.append(center[i][0])
        train.append(center[i][1])
    
    if len(train) < 8:
        for i in range(8 - len(train)):
            train.append(0)
            
    X.append(train)
    if choice == 0: 
        Y.append([1, 0, 0, 0])
    elif choice == 1:
        Y.append([0, 1, 0, 0])
    elif choice == 2:
        Y.append([0, 0, 1, 0])
    elif choice == 3:
        Y.append([0, 0, 0, 1])
```
