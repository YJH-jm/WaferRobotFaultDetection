```python
import csv
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import math
```

center 점 하나하나 plot

```python
for i in range(100):
        with open("fftdataok.csv", "rU") as file:
            data = csv.reader(file)
                #데이터 받아오기
            x = []
            for row in data:
                x.append(float(row[i+1]))
            x.pop(0)
            x = x[:484]

            result = []
                #의미있는 데이터 모으기
            for j in range(len(x)):
                if x[j] >= 0.4:
                    result.append([j, x[j]])

            if len(result) == 1: 
                model = KMeans(n_clusters = 1)
            elif len(result) == 2: 
                model = KMeans(n_clusters = 2)
            elif len(result) == 3: 
                model = KMeans(n_clusters=3)
            else:
                model = KMeans(n_clusters=4)
            model.fit(result)

            centers = model.cluster_centers_
                
               
            #dataShape(centers, choice, trainX, trainY)
        
           
            plt.scatter(x=centers[:,0], y=centers[:,1], marker='D', c='r')
            plt.xlim(0, 484)
            plt.ylim(0.4, 1)
            plt.xlabel('Frequency(Index)') 
            plt.ylabel('Magnitude') 
            plt.show()
```
